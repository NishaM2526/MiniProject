```python
import pandas as pd

df = pd.read_csv(r"../Data/Luxury_Housing_Bangalore.csv")
df
```

```python
df_copy = df.copy()
```

```python
df_copy
```

# Data Cleaning

```python
df_copy['Ticket_Price_Cr']=df_copy['Ticket_Price_Cr'].str.replace('Cr','')
df_copy['Ticket_Price_Cr']=df_copy['Ticket_Price_Cr'].str.replace('₹','')
df_copy['Ticket_Price_Cr']=df_copy['Ticket_Price_Cr'].astype(float)
```

```python
df_copy['Configuration']=df_copy['Configuration'].str.upper()
```

```python
df_copy['Micro_Market'] = df_copy['Micro_Market'].str.strip().str.title()
```

```python
df_copy.info()
```

```python
df_copy
```

```python
df_value=df_copy['Unit_Size_Sqft'].isna()
df_copy[df_value]
```

```python
df_copy[df_copy[['Unit_Size_Sqft','Ticket_Price_Cr','Amenity_Score']].isna().all(axis=1)]
```

```python
# drop columns with all null values in 'Unit_Size_Sqft' , 'Ticket_Price_Cr'  and 'Amenity_Score'
df_copy=df_copy.dropna(subset=['Unit_Size_Sqft','Ticket_Price_Cr','Amenity_Score'], how='all') 
```

```python
df_copy[df_copy[['Unit_Size_Sqft','Ticket_Price_Cr','Amenity_Score']].isna().all(axis=1)]
```

```python
df_copy[df_copy[['Unit_Size_Sqft','Ticket_Price_Cr']].isna().all(axis=1)]
```

```python
# drop rows with all null values in 'Unit_Size_Sqft' , 'Ticket_Price_Cr' 
df_copy=df_copy.dropna(subset=['Unit_Size_Sqft','Ticket_Price_Cr'], how='all') 
```

```python
df_copy[df_copy[['Unit_Size_Sqft','Ticket_Price_Cr']].isna().all(axis=1)]
```

```python
df_copy.isnull().sum()
```

```python
# Include mean/median values for 'Unit_Size_Sqft' and 'Ticket_Price_Cr' to fill the null values in the respective columns
df_copy=df_copy.fillna(
                        {   
                            'Unit_Size_Sqft':df_copy['Unit_Size_Sqft'].mean(),
                            'Ticket_Price_Cr':df_copy['Ticket_Price_Cr'].median(),
                            'Amenity_Score':df_copy['Amenity_Score'].median()
                        }
                      )
```

```python
df_copy.isnull().sum()
```

```python
df_copy['Buyer_Comments'].unique()
```

```python
# Fill the null values in 'Buyer_Comments' column with the previous non-null value using forward fill method
df_copy=df_copy.fillna({'Buyer_Comments': df_copy['Buyer_Comments'].ffill()})
```

```python
df_copy['Buyer_Comments'].unique()
```

```python
df_copy.isnull().sum()
```

```python
df_copy
```

```python
df_copy[df_copy.duplicated(keep=False)]
```

```python
df_copy[df_copy.duplicated()]
```

```python
# delete the duplicate rows and keep the first occurrence
df_copy=df_copy.drop_duplicates(keep='first')
```

```python
df_copy[df_copy.duplicated()]
```

```python
df_copy[df_copy.duplicated(keep=False)]
```

```python
df_copy
```

```python
# Calculate the price per squarefeet and create a new column 'Price_per_Sqft'
df_copy['Price_per_Sqft'] = ((df_copy['Ticket_Price_Cr']*10000000)/df_copy['Unit_Size_Sqft']).round(2)
```

```python
df_copy['Purchase_Quarter']=pd.to_datetime(df_copy['Purchase_Quarter'])
```

```python
# Extract the quarter number from the 'Purchase_Quarter' column and create a new column 'Quarter_Number'
df_copy['Quarter_Number']=df_copy['Purchase_Quarter'].dt.quarter
```
