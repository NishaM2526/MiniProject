```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

features = ['danceability','energy','loudness','speechiness',
            'acousticness','instrumentalness','liveness',
            'valence','tempo']

scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[features])
scaled_df = pd.DataFrame(scaled_features, columns=features)

for i in features:
    plt.figure(figsize=(8,4))
    sns.kdeplot(df[i], label='Original')
    sns.kdeplot(scaled_df[i], label='Scaled')
    plt.legend()
    plt.title(i)
    plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

features = ['danceability','energy','loudness','speechiness',
            'acousticness','instrumentalness','liveness',
            'valence','tempo']

scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[features])

decomposition = PCA(n_components=9)
output = decomposition.fit_transform(scaled_features)
print(np.cumsum(decomposition.explained_variance_ratio_))
output_df = pd.DataFrame(output,columns=['P1','P2','P3','P4','P5','P6','P7','P8','P9'])
output_df
```

```python
sns.scatterplot(output_df,x="P1",y="P2")
plt.show()
```
